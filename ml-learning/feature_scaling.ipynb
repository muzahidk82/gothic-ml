```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
df = pd.read_csv("./train.csv", usecols=["Age","Fare"])
```

```python
df.head()
```

```python
df["Age"].nunique()
```

```python
sc = StandardScaler()
df_new = pd.DataFrame(sc.fit_transform(df), columns= df.columns)
```

```python
df_new.head()
```

```python
pip install matplotlib
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
plt.title("Scatterplot before Standadization", fontsize=18)
sns.scatterplot(df, color='blue')
plt.subplot(1,2,2)
plt.title("Scatterplot after Standadization", fontsize=18)
sns.scatterplot(df_new, color="red")
plt.tight_layout()
plt.show()
```

$$
\text{Min Max Scaling} = \frac{x - \max(x)} {\max(x) - \min(x)}
$$

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler()
```

```python
df_new2 = pd.DataFrame(sc.fit_transform(df), columns=df.columns)
```

```python
df_new2.head(5)
```

```python
plt.figure(figsize=(14,7))
plt.subplot(1, 2, 1)
plt.title("Scatterplot before min_max scaling")
sns.scatterplot(df, color="red")
plt.subplot(1, 2, 2)
plt.title("Scatterplot after scaling")
sns.scatterplot(df_new2, color="blue")
plt.show()
```
